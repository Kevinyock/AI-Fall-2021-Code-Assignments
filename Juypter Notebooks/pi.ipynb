```python
import math
import random

# 10 ^ 3
# 10 ^ 4
# 10 ^ 5
# 10 ^ 6
n = [10**3,10**4,10**5,10**6]

listOfCalculatedPi = []
listOfErrorRates = []

dotsInCricle = set()
dotsInSquare = set()

def printOutput(n,calculatedPi,errorRate):
  n = "{:e}".format(n)
  calculatedPi = "{0:.6f}".format(calculatedPi)
  errorRate = "{0:.4f}".format(errorRate)
  print("n = ",n ,"pi = ",calculatedPi, "error =",errorRate,"%")

def generateRandomPoints():
  PointX = random.random()
  PointY = random.random()
  return (PointX, PointY)

def calculateErrorRate(calculatedPi):
  errorRate = abs(((math.pi - calculatedPi)/math.pi)*100)
  return errorRate

def generateListOfPoints(index):
  dotsInCricle.clear()
  dotsInSquare.clear()
  for val in range(n[index]):
    newPoint = generateRandomPoints()
    distance = math.sqrt(math.pow(newPoint[0],2) +(math.pow(newPoint[1],2)))
    if (distance <= 1):
      dotsInCricle.add(newPoint)
      dotsInSquare.add(newPoint)
    else:
      dotsInSquare.add(newPoint)

def calculateCalculatedPi(dotsInCricle,dotsInSquare):
  calculatedPi =  (4 * len(dotsInCricle))/len(dotsInSquare)
  return calculatedPi

```

```python
for val in range(len(n)):
  generateListOfPoints(val)
  CalcPi = calculateCalculatedPi(dotsInCricle,dotsInSquare)
  errorRate = calculateErrorRate(CalcPi)
  printOutput(n[val],CalcPi,errorRate)
```

```python
generateListOfPoints(1) #n = 10 ** 4
```

```python
# for n = 10 ** 4
import matplotlib.pyplot as plt
markersize = 5
zip_dotsInCricle = zip(*dotsInCricle)
zip_dotsInSquare = zip(*dotsInSquare.difference(dotsInCricle))
plt.scatter(*zip_dotsInSquare,marker='.',s=markersize, color = 'r')
plt.scatter(*zip_dotsInCricle,marker='.',s=markersize, color = 'b')
plt.title("Estimating Pi using Simulation")
plt.show()
```
